# TRAINING DATA
Creating multiple datapoints by splitting the total values of each column into many rows
1. 500 values per row - values selected in random order
2. 500 values per row - selected sequentially
3. Fixed rows - Random number of values between 0-500
4. Fixed rows - Random range of values

    start index = [random(0,500)]

```
import pandas as pd
import numpy as np
import random
```

```
trainingDataRandom60 = pd.read_parquet('drive/My Drive/training_raw_random_60.gzip')
trainingDataRandom60.head()
```

```
trainingDataSequential60 = pd.read_parquet('drive/My Drive/training_raw_sequential_60.gzip')
trainingDataSequential60.head()
```

    1. 500 random values - Random 60% training data
    2. 500 random values - Sequential 60% training data

```
training = []
```

```
len(trainingDataRandom60)
```

```
#1. 500 random values - Random 60% training data
for k in range(len(trainingDataRandom60)):
  rows = int(trainingDataRandom60['count'].values[k] / 500)
  for i in range(rows):
    random500Values = []
    for j in range(500):
      index = np.random.randint(0,trainingDataRandom60['count'].values[k])
      random500Values.append(trainingDataRandom60['Values'].values[k][index]) #pushing the random values to training dataset
    row = {'UniqueId':np.random.randint(491859, 550000), 'Values': random500Values, 'type':trainingDataRandom60['type'].values[k], 'count':len(random500Values)}
    training.append(row)
    random.shuffle(training)
trainDataRandom60Random500 = pd.DataFrame(training, columns = ['UniqueId', 'Values', 'type', 'count']) 
```
